#Statistical Methods

```python
#Seperating Data for analysis
legit=data[data.Class==0]
fraud=data[data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
#Will measure the statistics for legit transactions. (%ile)
```

```python
fraud.Amount.describe()
#Will measure the statistics for fraud transactions. (%ile)
```

```python
#Compare values of these transactions
data.groupby('Class').mean()
```

#Undersampling


Build a sample datastet from the original data set containing similar transactions of Legit & Fradulent Data

```python
# Since there are 492 Fraudulent Transactions, we will randomly select 492 legit transactions. And we will merge these two datasets.
# This way the dataset will become balanced & even distribution
legit_sample=legit.sample(n=492) #Random Sampling
```

```python
#Concat the legit_sample & fraud dataset
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
#Print first 10 values of the new dataset
new_dataset.head(10)
```

```python
#Print last 10 values of the new dataset
new_dataset.tail(10)
```

```python
#Check value count of new dataset
new_dataset['Class'].value_counts()
```

```python
#Find mean of new dataset
new_dataset.groupby('Class').mean()
```

The values haven't change much thus The nature of data has not changed as compared to the previous dataset. Thus this is a good sample.

```python
#Split the data into features & targets
X=new_dataset.drop(columns='Class', axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

#Splitting Data


```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
#Stratify=Y : evenly distributed 2 classes
#random_state=2 : a particular state
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model=LogisticRegression()
```

```python
# Training the Logistic regression model with training data
model.fit(X_train, Y_train)
```

```python
#Accuracy on Training Data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data : ', training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test Data : ', test_data_accuracy)
```
